#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reproducing a plot 1
np.random.seed(42) # 1.1 Set seed
students_affiliations = np.random.choice(["CS", "Engineering", "Math", "Physics"],size=1000) # Create a list of size 1000 from the categories and store in students_affiliations
_ = pd.Series(students_affiliations).value_counts().plot(kind='bar', rot=0, xlabel="Disciplines", ylabel="Counts", alpha=0.55, facecolor="#FF5733", title="Distribution of Students in the Program") # Plot using correct labels, color, alpha, and title
```

```python
random_data = np.random.normal(size=(1000, 2)) # Generate random data using normal distribution
pd.DataFrame(random_data).plot(kind='scatter', x=0, y=1, marker='s', xlabel='', ylabel='', title="Random Data Plot", alpha=0.55) # Scatter plot using squares
```
